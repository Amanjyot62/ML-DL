```python
def find_missin_values(dataset_nm):
    total_missing_cols=dataset_nm.isnull().sum().sort_values(ascending=False)[dataset_nm.isnull().sum().sort_values(ascending=False)!=0]
    percent_missing= round(dataset_nm.isnull().sum().sort_values(ascending=False)*100/len(dataset_nm),2)[dataset_nm.isnull().sum().sort_values(ascending=False)!=0]
    return pd.concat([total_missing_cols,percent_missing],axis=1,keys=['total_missing_cols','percent_missing'])


    
```

**Assumptions of Regression**

    Linearity -- Linear Regression should have presence of linearity between dependent and independent variable
    Homoscedasticity -- Variability of dependent variable doesnt increases as the value of independent variable increases
    Independence of Errors -- errors are independent of each other no pattern followed
    Multivariate Normality
    No Mutlicollinearity -- we expect there should not be presence of any kind of correlation between the dependent variables.

    Multicollinearity
No or Little multicollinearity: Multicollinearity is when there is a strong correlation between independent variables. Linear regression or multilinear regression requires independent variables to have little or no similar features. Multicollinearity can lead to a variety of problems, including:

The effect of predictor variables estimated by our regression will depend on what other variables are included in our model.

Predictors can have wildly different results depending on the observations in our sample, and small changes in samples can result in very different estimated effects.

With very high multicollinearity, the inverse matrix, the computer calculates may not be accurate.

We can no longer interpret a coefficient on a variable as the effect on the target of a one-unit increase in that variable holding the other variables constant. The reason behind that is, when predictors are strongly correlated, there is not a scenario in which one variable can change without a conditional change in another variable.

Heatmap is an excellent way to identify whether there is multicollinearity or not. The best way to solve multicollinearity is to use regularization methods like Ridge or Lasso.


```python
## Plot fig sizing. 
style.use('ggplot')
sns.set_style('whitegrid')
plt.subplots(figsize = (30,20))
## Plotting heatmap. 

# Generate a mask for the upper triangle (taken from seaborn example gallery)
mask = np.zeros_like(Train_Data.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True


sns.heatmap(Train_Data.corr(), 
            cmap=sns.diverging_palette(20, 220, n=200), 
            mask = mask, 
            annot=True, 
            center = 0, 
           );
## Give title. 
plt.title("Heatmap of all the Features", fontsize = 30);
```

```python
def SkewCorrecting(DATA):
    from scipy.stats import skew
    from scipy.special import boxcox1p
    from scipy.stats import boxcox_normmax
    Cont_features=list(DATA.select_dtypes('number').columns)
    Skewness=DATA[Cont_features].apply(lambda x: x.skew()).sort_values(ascending=False)
    high_skew = Skewness[abs(Skewness) > 0.5]
    skew_list_hgh=high_skew.index
    
    for high_f in skew_list_hgh:
        DATA[high_f]=boxcox1p(DATA[high_f], boxcox_normmax(DATA[high_f] + 1))
SkewCorrecting(Complete_data)       
```

```python
def overfit_reducer(df):
    """
    This function takes in a dataframe and returns a list of features that are overfitted.
    """
    overfit = []
    for i in df.columns:
        counts = df[i].value_counts()
        zeros = counts.iloc[0]
        if zeros / len(df) * 100 > 99.94:
            overfit.append(i)
    overfit = list(overfit)
    return overfit


overfitted_features = overfit_reducer(X)

X = X.drop(overfitted_features, axis=1)
X_sub = X_sub.drop(overfitted_features, axis=1)
```
